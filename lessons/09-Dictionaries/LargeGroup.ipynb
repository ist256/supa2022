```python
with open ("europe.json","r") as f:
    europe = json.load(f)
    
europe
```

```python
europe[8]['borders']
```

```python
def lookupname(europe, code):
    for country in europe:
        if country['alpha3Code'] == code:
            return country['name']
```

```python
lookupname(europe, 'GBR')
```

```python
namelist = []
for code in europe[8]['borders']:
    namelist.append(lookupname(europe, code))

namelist
```

```python
namelist = [ lookupname(europe, code) for code in europe[8]['borders'] ]
namelist
```
